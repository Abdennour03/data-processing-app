```python
Omega = {1, 2, 3, 4, 5, 6, 7, 8, 9}
A = {1, 2, 3}
B = {1, 5, 4, 3}
def P(event):
    return len(event) / len(Omega)

def Compliment(event):
    return 1 - P(event)

def Union(A, B):
    return A|B 

def Intersection(A, B):
    return A&B 
```
